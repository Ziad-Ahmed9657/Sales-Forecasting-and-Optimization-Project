```python
columns_to_plot = ['Sales', 'Profit', 'Quantity']

plt.figure(figsize=(18, 10))

for i, column in enumerate(columns_to_plot, 1):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    cleaned_data = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    df.drop(outliers.index, inplace=True)

    plt.subplot(2, 3, i)
    sns.boxplot(df[column])
    plt.title(f'{column} (Before)')

    plt.subplot(2, 3, i + 3)
    sns.boxplot(cleaned_data[column])
    plt.title(f'{column} (After)')

plt.suptitle('Boxplots Before and After Handling Outliers', fontsize=16)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

Q1 = df["Discount"].quantile(0.25)
Q3 = df["Discount"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["Discount"] < lower_bound) | (df["Discount"] > upper_bound)]
cleaned_data = df[(df["Discount"] >= lower_bound) & (df["Discount"] <= upper_bound)]
df.drop(outliers.index, inplace=True)

plt.subplot(1, 2, 1)
sns.boxplot(df["Discount"])
plt.title('Discount (Before Handling Outliers)')

plt.subplot(1, 2, 2)
sns.boxplot(cleaned_data["Discount"])
plt.title('Discount (After Handling Outliers)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

Q1 = df["Shipping Cost"].quantile(0.25)
Q3 = df["Shipping Cost"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["Shipping Cost"] < lower_bound) | (df["Shipping Cost"] > upper_bound)]
cleaned_data = df[(df["Shipping Cost"] >= lower_bound) & (df["Shipping Cost"] <= upper_bound)]
df.drop(outliers.index, inplace=True)

plt.subplot(1, 2, 1)
sns.boxplot(df["Shipping Cost"])
plt.title('Shipping Cost (Before Handling Outliers)')

plt.subplot(1, 2, 2)
sns.boxplot(cleaned_data["Shipping Cost"])
plt.title('Shipping Cost (After Handling Outliers)')

plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df, x="Category", y="Sales", estimator='mean')
plt.title('Average Sales per Category')
plt.xticks(rotation=45)
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

sns.barplot(data=category_sales, x="Category", y="Sales")
plt.title('Total Sales per Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(df["Category"], df["Sales"], alpha=.3)
plt.xticks(rotation=50)
plt.show()
```

```python
categories = df["Category"].unique()
colors = plt.cm.rainbow(np.linspace(0, 1, len(categories)))

for category, color in zip(categories, colors):
	subset = df[df["Category"] == category]
	plt.scatter(subset["Category"], subset["Sales"], alpha=.3, label=category, color=color)

plt.legend()
plt.show()
```
